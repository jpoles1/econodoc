# Collecting Medicare DRG Payment Data

Using data from: https://data.cms.gov/Medicare-Inpatient/National-Summary-of-Inpatient-Charge-Data-by-Medic/efwk-h4x3

```python
import pandas as pd
```

```python
raw_data = pd.read_csv("data/National_Summary_of_Inpatient_Charge_Data_by_Medicare_Severity_Diagnosis_Related_Group__MS-DRG___FY2015.csv")
raw_data.head(10)
```

```python
clean_data = raw_data.copy(deep=True)
clean_data["DRG Code"] = [x.split(" -")[0] for x in clean_data["DRG Definition"]]
clean_data["DRG Definition"] = [x.split("- ")[1] for x in clean_data["DRG Definition"]]
clean_data.head(10)
```
